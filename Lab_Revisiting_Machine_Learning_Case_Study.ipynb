### Identify columns that over 85% missing values

```python
# creating a function that makes a dataframe with name of cols and more than 85% missing values


def nulls_high_perc(categorical, perc=0.85):
    high_perc_df = pd.DataFrame(categorical.isna().sum() / len(categorical)).reset_index()
    high_perc_df.columns = ['column_name', 'nulls_percentage']
    cols_above_threshold = high_perc_df[high_perc_df['nulls_percentage'] > perc]
    drop_list_two = list(cols_above_threshold['column_name'])
    return drop_list_two
```

```python
drop_list_two = nulls_high_perc(categorical)
drop_list_two
```

```python
nulls = pd.DataFrame(categorical.isna().sum()*100/len(categorical), columns=['percentage'])
nulls_85= nulls[nulls['percentage'] >= 85].sort_values('percentage', ascending=False)
nulls_85
```
